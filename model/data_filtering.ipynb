```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Veriyi yükleyin
data = pd.read_csv('2024-06-10.csv')

# Sektörleri oluşturun
sectors = ['Technology', 'Finance', 'Healthcare', 'Energy', 'Consumer Goods', 'Utilities', 'Telecommunications', 'Materials', 'Real Estate']

# Portföy oluşturun
unique_tickers = data['Ticker'].unique()
portfolio_tickers = random.sample(list(unique_tickers), 15)
portfolio = pd.DataFrame({
    'Ticker': portfolio_tickers,
    'Sector': [random.choice(sectors) for _ in range(len(portfolio_tickers))]
})

# Portföyü CSV olarak kaydedin
portfolio.to_csv('Portfolio_1.csv', index=False)

# İşlemler oluşturun
transactions = []
for _ in range(500):
    ticker = random.choice(portfolio_tickers)
    date = datetime.now() - timedelta(days=random.randint(1, 365))
    quantity = random.randint(1, 1000)
    price = data.loc[data['Ticker'] == ticker, 'Close'].iloc[0]
    transaction_type = random.choice(['Alış', 'Satış'])
    transactions.append([date, ticker, quantity, price, transaction_type])

# İşlemleri DataFrame'e dönüştürün ve CSV olarak kaydedin
transactions_df = pd.DataFrame(transactions, columns=['Tarih', 'Hisse Adı', 'Adet', 'Fiyat', 'İşlem Tipi'])
transactions_df.to_csv('islemler.csv', index=False)
```
